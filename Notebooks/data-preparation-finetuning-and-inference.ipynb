```python
import re
import pandas as pd

def parse_questions_a(content, filename):
    # Split the content into lines
    lines = content.split('\n')
    
    # Initialize lists to store question and answer pairs
    questions = []
    answers = []
    
    # Regular expressions to match questions and answers
    # question_pattern = re.compile(r'^\*\*Q\d+:\*\*\s(.+)$')
    # answer_pattern = re.compile(r'^\*\*A\d+:\*\*\s(.+)$')
    # question_pattern = re.compile(r'^\*\*\d+\.\sQ:\s(.+)\*\*$')
    # answer_pattern = re.compile(r'^A:\s(.+)$')
    # question_pattern = re.compile(r'^\*\*\d+\.\s(.+?)\*\*$')
    # answer_pattern = re.compile(r'^Answer:\s(.+)$')
    question_pattern = re.compile(r'^\*\*Q\d+:\*\*\s(.+)$')
    answer_pattern = re.compile(r'^\*\*A\d+:\*\*\s(.+)$')
    
    # Temporary variables to store the current question and answer
    current_question = None
    current_answer = None
    
    # Iterate through lines to extract questions and answers
    for line in lines:
        # Check if the line matches a question
        question_match = question_pattern.match(line)
        if question_match:
            # Store the question text
            current_question = question_match.group(1).strip()
            continue
        
        # Check if the line matches an answer
        answer_match = answer_pattern.match(line)
        if answer_match:
            # Store the answer text
            current_answer = answer_match.group(1).strip()
        
        # If both question and answer are available, append them
        if current_question and current_answer:
            questions.append(current_question)
            answers.append(current_answer)
            # Reset temporary variables for the next pair
            current_question = None
            current_answer = None
    
    # Create a DataFrame
    df = pd.DataFrame({
        # 'filename': [filename] * len(questions),
        'question': questions,
        'answer': answers
    })
    
    return df
```

```python
import re

def parse_questions(content, filename):
    # Split the content into lines
    lines = content.split('\n')
    
    # Initialize lists to store questions and answers
    questions = []
    answers = []
    
    # Regular expression to match numbered questions
    question_pattern = re.compile(r'^\d+\.\s(.+)$')
    
    # Iterate through lines to extract questions
    for line in lines:
        match = question_pattern.match(line)
        if match:
            questions.append(match.group(1).strip())
            answers.append("")  # Initialize with empty answer
    
    # Create a DataFrame
    df = pd.DataFrame({
        'filename': [filename] * len(questions),
        'question': questions,
        'answer': answers
    })
    
    return df
```

```python
def clean_question(question):
    import re
    
    # Remove "**Q:" from front and "**" from back
    cleaned_question = re.sub(r'^(\*{1,2})?Q[:.]\s*', '', question).rstrip('*').strip()
    return cleaned_question
```

```python
def generate_answers_once_groq(file_path, questions):
    # Read file content
    with open(file_path, 'r') as file:
        file_content = file.read()
    
    # Enhanced prompt with multiple questions
    prompt = f"""
    You are an expert assistant specializing in the Insurance domain. Your task is to analyze the provided insurance 
    document and generate accurate, short, and concise answers to the given questions.

    For each answer:
    1. Be concise, fact-based, and derived solely from the document's content.
    2. Avoid directly referencing the document's structure, headings, or sections.
    3. Treat the document's information as general knowledge and frame the response professionally.
    4. Be specific, addressing the question clearly and completely without including irrelevant details.
    5. The scope of answers must be only the document content.
    
    Here is the insurance document content:
    {file_content}

    Please answer the following questions:
    {' '.join([f'Q{i+1}: {q}' for i, q in enumerate(questions)])}
    
    Provide your answers in the following format:
    A1: [Your answer to Q1]
    A2: [Your answer to Q2]
    ...
    """
    
    chat_completion = g_client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        model="llama-3.3-70b-versatile",
    )
    
    # Extract and parse the answers
    response_content = chat_completion.choices[0].message.content
    answers = response_content.split('\n')
    
    # Create a dictionary to store question-answer pairs
    qa_pairs = {}
    for answer in answers:
        if answer.startswith('A'):
            parts = answer.split(': ', 1)
            if len(parts) == 2:
                question_number = int(parts[0][1:])  # Extract the number from 'A1', 'A2', etc.
                qa_pairs[questions[question_number-1]] = parts[1].strip()
    
    return qa_pairs

# Example usage:
file_path = '/kaggle/input/synthetic-insurance-data/business-guardian.md'
for file_p in question_files_list:
    df = pd.read_csv(file_p)
    questions = [clean_question(row['question']) for _, row in df.iterrows()]
    # print(questions)
    
answers = generate_answers_once_groq(file_path, questions)

# Print the results
for question, answer in answers.items():
    print(f"Q: {question}")
    print(f"A: {answer}")
    print()
```

```python
import pandas as pd
import json


# Function to transform DataFrame to JSONL format
def dataframe_to_jsonl(df, output_file):
    with open(output_file, "w") as f:
        for _, row in df.iterrows():
            messages = [
                {"role": "user", "content": row["question"]},
                {"role": "assistant", "content": row["answer"]},
            ]
            json_record = {"messages": messages}
            f.write(json.dumps(json_record) + "\n")
```
